### EJERCICIO

La columna **budget** corresponde al **Presupuesto** y la columna **gross** corresponde a los **ingresos**, queremos agregar una columna al DataFrame donde calculemos las **ganancias**. Llámela ganancias y agréguela a df_movies.

Para finalizar guarde las columnas movie_title, director_name, budget, gross y ganancias como df_money y muestre los 5 primeros elementos 

```python
# Escriba aquí su código
df_movies['ganancias']=df_movies.gross - df_movies.budget
df_movies
df_money= df_movies[['movie_title','director_name','budget','gross','ganancias']].copy()
df_money.head(5)
```

### EJERCICIO
A partir de df_money muestre cuáles son las 10 películas con mayores ganancias? Guárdelo en df_topten_movies y muestre los resultados

```python
# Escriba aquí su código
df_topten_movies=df_money.nlargest(10,'ganancias')
df_topten_movies
```

### EJERCICIO

En un gráfico de Barras Verticales o Columnas, muestre las ganancias de cada una de estas  películas top ten

```python
# Escriba aquí su código

df_topten_movies.plot(kind='bar',x='movie_title',y='ganancias',figsize=[10,8], grid=True)
```

### EJERCICIO

Algunos directores han dirigido varias películas y cada una de ellas ha obtenido distintas ganancias. Queremos saber cuántas ganancias en total tuvo cada director. Hágalo a partir del df_money. Llame al resultado directores.

```python
# Escriba aquí su código

directores=df_movies.groupby('director_name').sum().ganancias
directores
```

```python
# Escriba aquí su código
directores.nlargest(10)# se ordenó el resultado para comprobar que estaba bien mostrando los 10 primeros
```

### EJERCICIO
Ahora queremos ordenar la lista anterior de mayor a menor. Tenga cuidado, posiblemente directores no es un DataFrame, sino una Serie, si es así debería usar sort_values ... 
Guarde el resultado ordenado con el nombre directores2 y muéstrelo

```python
# Escriba aquí su código
directores2= directores.sort_values(ascending=False)
directores2
```

Muestre un listado con las películas que ha dirigido el director que tiene las **mayores ganancias promedio** (muestre todas las columnas de df_money).
